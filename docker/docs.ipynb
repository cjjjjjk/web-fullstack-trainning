# **DOCKER**
### **DOCKER**
**Docker** - Nền tảng giúp build, deploying, running ứng dụng dễ dàng hơn bằng cách xây dựng các containers dựa trên nền tảng ảo hoá.  
**Docker** thực hiện đóng gói phần mềm thành các container, chứa đựng tất cả thông tin cần thiết để ứng dụng hoạt động như: thư viện, công cụ, giúp đảm bảo ứng dụng chạy đồng nhất trên mọi môi trường. Cho phép chạy ứng dụng trên các môi trường, thiết bị khác nhau.  
**Docker container**: Gói phần mềm chứa tất các các thành phần cần thiết của một ứng dụng: mã nguồn, thư viện,... đản bảo ứng dụng chạy đồng nhất trên mọi môi trường, đặc điểm:  
-   Dễ ứng dụng: 'build once, run anywhere'
-   Nhẹ và nhanh: có thế chạy trong vài giây
-   Khả năng mở rộng linh hoạt: có thể chia nhỏ chức năng ứng dụng thành các container riêng lẻ: database container, ... Docker có khả năng liên kết container linh hoạt.
-   Hiệu suất cao: Chia sẻ cùng một hệ điều hành kernel, giảm thiểu tài nguyên hệ thống so với các máy ảo.
-   Quản lý phụ thuộc: Container quản lý tất cả phụ thuộc của ứng dụng trong 1 container, đảm bảo không có sự mâu thuẫn.


### **CÁC KHÁI NIỆM**
- Docker client: Giao diện tương tác với API gửi lệnh đến Docker Deamon.
- Docker deamon: server Docker cung cấp API, quản lý images, container, networks, volumne.
- Docker volumes: Lưu dữ liệu liên tục bền vứng cho việc sử dụng và tạo apps. Dữ liệu trong container mất đi khi bị xoá, nhưng dữ liệu trong volumes không mất đi. Nhằm lưu database, ảnh upload,...
- Docker images: Mẫu template chứa toàn bộ mã ứng dụng, thư viện,... được build từ docker file
- Docker network: kết nối các container lại với nhau, trên 1 host hoặc nhiều host.
- Docker Files: File định nghĩa cách build ra một docker images.  
- Docker compose: công cụ chạy ứng dụng với nhiều docker container một cách dễ dàng. Định nghĩa run-multi container cho docker.
- Docker Hub: Registry lớn nhất của docker images.

Dockerfile  `build` → Docker Image  `run` → Docker container

### **Dockerfile**
File config cho docker để build ra images. Định nghĩa cách build ra các images.  
Docker file sử dụng image cơ bản để xây dựng lớp image ban đầu, các lớp bổ sung xêp chồng lên lớp cơ bản này sau đó.  
*Các config trong dockerfile*  
| config | chức năng |
| --- | --- |
| FROM | chỉ định image gốc |
| LABEL | cung cấp metadata cho image, thông tin maintainer |
| ENV | biến môi trường  |
| RUN | Tạo lệnh thực thi khi build image |
| COPY | sao chép các file từ máy chủ và thư mục vào container |
| ADD | như copy, hỗ trợ tự động giải nén /.tar |
| WORKDIR | thiết lập thư mục đang làm việc |
| VOLUME | định nghĩa thư mục lưu trữ dữ liệu cho image |
| CMD | cung cấp tập lệnh mặc định chạy khi container start, bị ghi đè |
| ENTRYPOINT | lệnh cấu hình chương trình chạy khi container start, không bị ghi đè |



```sh
# dockerfile xây dựng một dockerimage cho ứng dụng python. 

# xác định image cơ sở
FROM python:3.12-slim
# metadata                           
LABEL maintainer="Haihv <haihv@hai.com>"

# biến môi trường và chỉ định thư mục làm việc
ENV APP_DIR=/app
WORKDIR $APP_DIR

# sao chép thư mục hiện tại sang thư mục làm việc container /app
COPY . .
# lệnh thực thi khi xây dựng dockerimage, cài đặt gói thư viện trong lib.txt
RUN pip install -r lib.txt

# container lắng nghe ở cổng 8000
EXPOSE 8000
# thư mục chứa dữ liệu bền vững - volume của container
VOLUME ["/data"]

# lệnh thực thi khi container chạy + đối số mặc định, kết quả thực thi: python main.py
ENTRYPOINT ["python"]
CMD ["main.py"]

```

### **Docker Compose**
Công cụ dùng để định nghĩa run multi-container cho ứng dụng docker. Sử dụng file yaml để config các services cho ứng dụng. Docker compose dùng để build và run các container, các thao tác tương tự nhu docker run.  

Mục tiêu: Khởi chạy tất cả container cần thiết cho ứng dụng hoạt động. 
```yml 
# docker-componse.yaml cho nodejs + mongo
version: '3'
services:
  web:
    image:      # docker hub image
    build: .    # chỉ định build image từ docker file
        context: .
        dockerfile: Dockerfile
    ports:
      - "3000:3000"
    volumes:
      - .:/app
    depends_on:
      - mongo

  mongo:
    image: mongo:6
    volumes:
      - mongo-data:/data/db

volumes:
  mongo-data:
```

*Các lệnh thao tác docker compose*  
| lệnh | tác vụ | 
| --- | --- |
| `docker-compose up` | chạy docker compose và các service được cấu hình trong docker-compose.yaml|
| `docker-compose down` | xoá container, network, volume |

### **Tối ưu hoá kích thước Docker Image**

```sh
FROM node:20

WORKDIR /app

COPY package*.json ./
RUN npm install             #  cài cả devDependencies

COPY . .                    #  copy toàn bộ source (kể cả node_modules, .git nếu không có .dockerignore)

RUN npm run build           #  build ngay trong image, giữ lại mọi thứ

CMD ["node", "dist/main"]

# Kết quả: 2.22 GB kích thước image
# Tối ưu hoá kích thước image còn 377MB
FROM node:20-alpine AS builder              # Phần build
WORKDIR /app

COPY package*.json ./
RUN npm install

COPY . .
RUN npm run build

FROM node:20-alpine                         # Tách biệt phần runtime
WORKDIR /app

COPY --from=builder /app/dist ./dist        # copy phần được build trong dist
COPY --from=builder /app/package*.json ./   # phần cần thiết bên ngoài

RUN npm install --only=production           # cài dependencies cho production only

CMD ["node", "dist/main"]

```

### **Docker network**
network trong Docker là phương pháp để các container giao tiếp với nhau, hoặc với internet, host ... Docker network Là một lớp trừu tượng giúp các container có thể:
- Giao tiếp nội bộ container-to-container
- Giao tiếp với host (host-to-container)
- Giải tiếp với internet
- Áp dụng chính sách mạng khi giao tiếp.

Các network mặc định không được phép xoá:
- **`host`** : mạng host 
- **`bridge`**: network mặc định container
- **`none`**: không gắn network nào 

Các loại driver được cung cấp:  
| Tên driver  | Mô tả ngắn gọn                                                                                    |
| ----------- | ------------------------------------------------------------------------------------------------- |
| **bridge**  | Mạng mặc định khi bạn chạy `docker run`. Các container trong cùng bridge có thể giao tiếp nội bộ. |
| **host**    | Container sử dụng mạng của host, không có lớp mạng riêng.                                         |
| **null**    | Không cấu hình mạng cho container.                                                                |
| **overlay** | Dùng cho Docker Swarm, kết nối container giữa nhiều máy chủ Docker khác nhau.                     |
| **macvlan** | Gán địa chỉ MAC riêng cho container, như một thiết bị vật lý thực thụ trong mạng LAN.             |
| **custom**  | Bạn có thể tạo mạng bridge/overlay riêng để kiểm soát tốt hơn về DNS, alias, subnet,...           |
