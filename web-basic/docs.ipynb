### **1. HTML / JS**  
HTML: Ngôn ngữ đánh dấu - tạo cấu trúc và nội dung trang web.(HTML - Hyper Text Markup language)    
*Cấu trúc cơ bản của tài liệu HTML*  
  

**DOM - Document Object Model**: Mô hình đối tượng tài liệu HTML, giao diện lập trình thể hiện cấu trúc trang web, các thẻ element được biểu diễn dưới dạng cây đối tượng và thẻ html là nút gốc.  

**JS / TS**
JS(Javasctip): Ngôn ngữ lập trình web, xử lý sự kiện. Được phát triển dựa trên cách đặt tên của Java, là ngôn ngữ lập trình nguyên mẫu được phát triển bằng C.
TS(Typescript): Ngôn ngữ lập trình mở rộng của JS, bổ sung cú pháp lập trình định kiểu tĩnh và hướng đối tượng. Chỉ tồn tại trong quá trình biên dịch. Và cần có Typescript Compiler để chuyển sang JS (Trình duyệt chỉ hiểu được HTML, CSS, JS)  
*Về mặt Runtime, hai ngôn ngữ là như nhau* 

**Các tính năng TS**
- NULL checking: Ngăn chặn lỗi khi biến có giá trị null.
- Generics: Định kiểu dữ liệu mà không xác định trước kiểu cụ thể -> đảm bảo tính an toàn khi xây dựng các phương thức,.. tái sử dụng mà đáp ứng kiểu dữ liệu an toàn. 
- Access Modifiers: Cung cấp giới hạn truy cập cho thuộc tính và phương thức lớp  
- Static typing, Type Annotations: Gán kiểu tĩnh cho biến, tham số. Kiểu của biến được xác định trong quá trình biên dịch, tại thời điểm biên dịch, phát hiện lỗi trước khi chạy mã.   

**Nhược điểm**: Thiết lập cồng kềnh, chỉ là phần ngôn ngữ mở rộng hỗ trợ.

### **2. CSS**
CSS: Ngôn ngữ thiết kế (CSS - Cascading Style Sheets) dành cho tài liệu HTML, XML  
*Cách triển khai:*  
| Cách triển khai | Mô tả|
| -- | -- |
| External stylesheet | Tách riêng biệt file thiết kế | 
| Internal stylesheet | Thiết kế bên trong thẻ style | 
| Inline stylesheet| Thiết kế trực tiếp cho từng phần tử qua thuộc tính style |   

*CSS box model*  
  


###  **3.CSS NÂNG CAO**
- Grid Layout: Triển khai layout 2 chiếu (hàng + cột)
- Flexbox: Triển khai layout 1 chiều (hàng hoặc cột)
- CSS Variable: Khả năng tái sử dụng màu, tăng khả năng chỉnh sửa  
- Responsive Design: Hiển thị thông tin tốt + đầy đủ ở trên nhiều màn hình thiết bị khác nhau  
*CSS Media query*  
  
- Advanced Selectors: Select các thẻ một cách linh hoạt
- CSS Proprocessors: Sass, LESS
- CSS Module: Phân chia module css
- CSS Performance Optimization: 
    - Không sử dụng quá nhiều !important
    - Không select quá sâu, chỉ nên tối đa ở 3 cấp select

### **3. BOOSTRAP / TAILWIND CSS**
Bootstrap và Tailwind CSS là framework CSS mã nguồn mở miễn phí.   

*So sánh tailwind và bootstrap*

| **Tiêu chí**                  | **Tailwind CSS**                                  | **Bootstrap**                                     |
| ------------------------- | ------------------------------------------------- | ------------------------------------------------- |
| Nguyên tắc     | Utility-first (dùng class nhỏ, kết hợp linh hoạt) | Component-based (dựng sẵn) |
| Tùy biến giao diện    | Rất linh hoạt, tự thiết kế từ đầu                 | Hạn chế, không linh hoạt    |
| Animation | Cung cấp cú pháp :hover, duration, transition | Không được cung cấp các cú pháp xử lý animation, hover,... |
| Animation & hiệu ứng  | Hỗ trợ tốt, dễ custom, dễ kết hợp thư viện ngoài  | Cơ bản , khó mở rộng    |
| Tốc độ phát triển UI  | Nhanh khi thành thạo, mất nhiều thời gian để làm quen       | Nhanh nhờ sử dụng UI dựng sẵn                        |
| Kích thước tệp CSS khi build     | Nhẹ                           | Nặng do conponent-based               |
| Responsive            | Tốt      | Tốt                            |
|  | Phát triển mạnh, nhiều, sáng tạo                              | Lớn, lâu đời, tài liệu nhiều                  |

**Tổng kết**  
**BS5** phù hợp với các dự án không đặt nặng vào giao diện phức tạp, cần triển khai nhanh và có một thư viện UI chuẩn sẵn có và đầu đủ.  
*VD: Hệ thống nội bộ cho doanh nghiệp, web admin,...*     

**Tailwind CSS** có sự sáng tạo, xây dựng linh hoạt, phù hợp dự án yêu cầu linh hoạt UI và tuỳ chỉnh cao, yêu cầu thiết kế độc đáo.  
*VD: Các trang thương mại, quảng bá sản phẩm,...*


*So sánh cú pháp Tailwind CSS và Bootstrap*
| Tính năng                    | Tailwind CSS                               | Bootstrap 5 (BS5)                                         |
| ---------------------------- | ------------------------------------------ | --------------------------------------------------------- |
| **Màu nền**                  | `bg-red-500`                               | `bg-danger`                                               |
| **Màu chữ**                  | `text-white`, `text-gray-700`              | `text-white`, `text-muted`, `text-dark`                   |
| **Căn (text)**          | `text-center`                              | `text-center`                                             |
| **Lề (margin)**              | `mt-4`, `mb-2`, `mx-3`, `my-1`             |  Tương tự                                                 |
| **Padding**                  | `pt-2`, `px-4`, `py-1`                     | `pt-2`, `px-4`, `py-1`                                    |
| **Flexbox**                  | `flex`, `justify-between`, `items-center`  | `d-flex`, `justify-content-between`, `align-items-center` |
| **Grid hệ thống**            | `grid grid-cols-3 gap-4`                   | `row`, `col-4`, `g-3`                                     |
| **Hiển thị ẩn/hiện**         | `hidden`, `block`, `md:block`              | `d-none`, `d-block`, `d-md-block`                         |
| **Bo góc**                   | `rounded`, `rounded-lg`, `rounded-full`    | `rounded`, `rounded-lg`, `rounded-circle`                 |
| **Đổ bóng (shadow)**         | `shadow`, `shadow-lg`, `shadow-md`         | `shadow`, `shadow-lg`                                     |
| **Kích thước chữ**           | `text-sm`, `text-lg`, `text-xl`            | `fs-6`, `fs-4`, `fs-1`                                    |
| **Chiều rộng/cao cụ thể**    | `w-1/2`, `h-32`, `w-full`                  | `w-50`, `h-100` (theo pixel, %), hoặc custom              |
| **Responsive** (VD: padding) | `md:p-4`, `lg:pt-6`                        | `p-md-4`, `pt-lg-6`                                       |
| **Button**                   | `bg-blue-500 text-white px-4 py-2 rounded` | `btn btn-primary`                                         |
| **Hover** | `hover:bg-blue-500` | Không có |

Có thể thấy, tailwind css có cú pháp triển khai phức tạp và linh hoạt hơn, triển khai màu, kích thước,..  linh động và đa dạng hơn so với bs5.

### **5. NODEJS**
**Nodejs** - Nên tảng cho phép chạy JS ngoài trình duyệt (phía máy chủ), ngôn ngữ chính: Javascript.  
**Mục tiêu** : Phát triển ứng dụng web phía máy chủ, hoặc các ứng dụng real-time, nhiều thao tác đọc ghi dữ liệu.  

Nodejs là đơn luồng, hỗ trợ bất đồng bộ nhớ **Event Loop** và **livbuv**, cho phép xử lý song song tác vụ mà không cần thêm luồng. Các tác vụ không đảm bảo tg kết thúc được đưa vào **Thread pool(livbuv)**, sau khi xong, kích hoạt callback.
